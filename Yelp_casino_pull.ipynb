### Terminal Code for Yelp API, Casino Search

```python
#!/bin/bash

# Yelp API Key
API_KEY="yourkey"

# Define cities array
cities=(
'HighlandsRanch,CO'
'Longmont,CO'
'Centennial,CO'
'Sparks,NV'
'Westminster,CO'
'Boulder,CO'
'Arvada,CO'
'TheWoodlands,TX'
'Billings,MT'
'Thornton,CO'
'Greeley,CO'
'Lakewood,CO'
'Paradise,NV'
'SunriseManor,NV'
'Enterprise,NV'
'Tacoma,WA'
'SpringValley,NV'
'Bremerton,WA'
'NorthLasVegas,NV'
'Henderson,NV'
'Aurora,CO'
'Reno,NV'
'Denver,CO'
'Seattle,WA'
)

# Loop through each city in the array
for city in "${cities[@]}"; do        
        curl --request GET \
             --url "https://api.yelp.com/v3/businesses/search?term=casino&location=$city&sort_by=best_match&limit=50&offset=20&categories=casinos" \
             --header "Authorization: Bearer $API_KEY" \
             --header "accept: application/json" > "${city// /}_casino50.json"

        echo "Data for $city with offset $offset saved to ${city// /}_casino50.json"
    done
done
```

```python
#combining all .json files into one csv
import pandas as pd
import glob
import os
import json

def read_json_file(filepath):
    try:
        with open(filepath, 'r') as file:
            data = json.load(file)
        # Assuming 'businesses' is the key containing the list of business dictionaries
        if 'businesses' in data:
            return pd.json_normalize(data['businesses'])
        else:
            print(f"No 'businesses' key found in {filepath}")
            return None
    except Exception as e:
        print(f"Error reading {filepath}: {e}")
        return None

directory = 'yourdirectory'
json_files = [f for f in os.listdir(directory) if f.endswith('.json')]

dataframes = []
for json_file in json_files:
    df = read_json_file(os.path.join(directory, json_file))
    if df is not None:
        dataframes.append(df)

if dataframes:
    casino2 = pd.concat(dataframes, ignore_index=True)
    print("Successfully concatenated DataFrames")
else:
    print("No valid DataFrames to concatenate")
```

```python
## Flattening json files
import json
import glob
import pandas as pd

# Function to extract total businesses from a Yelp API JSON response
def extract_total_businesses(json_data):
    return json_data.get('total', 0)

# List to store data for DataFrame
data_list = []

# List all JSON files in the 'Casino_pull' directory
json_files = glob.glob('Casino_pull/*.json')

# Iterate over each file
for file in json_files:
    with open(file, 'r', encoding='utf-8') as f:
        data = json.load(f)
        total_businesses = extract_total_businesses(data)
        data_list.append({
            'File': file,
            'Total Businesses': total_businesses
        })

# Create DataFrame
df = pd.DataFrame(data_list)

# Display DataFrame
print(df)
```
